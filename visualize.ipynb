```python
# %matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_vectors(v1, v2, v3):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # 绘制向量
    vectors = {"v1": v1, "v2": v2, "v3": v3}
    for name, v in vectors.items():
        ax.quiver(0, 0, 0, v[0], v[1], v[2], label=name, length=1.5)  # Adjust the length parameter to make the vectors larger
        ax.text(v[0], v[1], v[2], name, color='red')

    # 设置图形的边界
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])

    # 显示图例
    ax.legend()

    plt.show()
```

```python
# (compute)wo: 0.480018 -0.660018 0.577891
# (compute)wi: -0.544870 0.181841 -0.818566
# (compute)wh: 0.407175 0.467499 0.784635
wo = np.array([0.480018, -0.660018, 0.577891])
wi = np.array([-0.544870, 0.181841, -0.818566])
wh = np.array([0.407175, 0.467499, 0.784635])
plot_3d_vectors(wo, wi, wh)
x = np.array([1, 0, 0])
y = np.array([0, 1, 0])
z = np.array([0, 0, 1])
plot_3d_vectors(x, y, z)
```
